```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_excel('data.xls')
```

```python
df.head()
```

```python
df = pd.read_excel('data.xls', header = None)
df.head()
```

```python
records = []
for i in range(0, 22):
    records.append([str(df.values[i,j]) for j in range(0, 6)])
```

```python
association_rules = apriori(records, min_support=0.09, min_confidence=0.5, min_lift=1.2, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
association_results[0]
```

```python
results = []
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    value0 = str(items[0])
    value1 = str(items[1])
    value2 = str(item[1])[:6]
    value3 = str(item[2][0][2])[:6]
    value4 = str(item[2][0][3])[:6]
    
    rows = (value0, value1, value2, value3, value4)
    
    results.append(rows)
    
    label = ['title1'], ['title2'],['support'], ['confidence'],['lift']
    
    store_suggestion = pd.DataFrame.from_records(results,columns=label)
    
    print(store_suggestion)
```
